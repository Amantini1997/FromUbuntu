```python
import pandas as pd
import numpy as np

dfq = pd.read_csv("./survey_questions_meta.csv")

dfq.head()
```

```python
dfr = pd.read_csv("./survey_responses.csv")

dfr.head()
```

```python
undergrads = dfr[dfr["Career"] == "UGRD"]
```

```python
undergrads
```

```python
DEPARTMENT_INDEX = 7
STRESS_INDEX = 13
SATISFACTION_INDEX = 14

departments = set(dfr.iloc[:, DEPARTMENT_INDEX])
for dep in departments:
    
    dep_rows = dfr[dfr.iloc[:, DEPARTMENT_INDEX] == dep]
    students_count = sum(dep_rows)
    
    stress_col = dep_rows.iloc[:, STRESS_INDEX]  
    stress_avg = stress_col.sum() / students_count
    
    satisfaction_col = dep_rows.iloc[:, SATISFACTION_INDEX]
    satisfaction_avg = satisfaction_col.sum() / students_count

    print "DEP: ", dep, "  STRESS: ", stress_avg, "   SATIS: ", satisfaction_avg 

    
```
